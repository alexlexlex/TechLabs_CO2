```
#import libraries
%matplotlib inline
import numpy as np
from sklearn.linear_model import LinearRegression
import pandas as pd
import matplotlib.pylab
import matplotlib.pyplot as plt
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 20, 16
import warnings
import itertools
warnings.filterwarnings("ignore") # specify to ignore warning messages

#read excel file while reading it as a timeseries dataset
df = pd.read_excel(r'/content/Pop.xlsx', parse_dates=['Year'], index_col='Year')
```

```
#Population and Emissions
plt.plot(df['Population'], df['CO2 Emissions (tons)'])
plt.xlabel("Population (Billion)", fontsize=20)
plt.ylabel("CO2 Emissions (tons)", fontsize=20)
plt.title("Population & CO2 Emissions Growth", fontsize=20)
```

```
#Population change and Emissions change
plt.plot(df['CO2 emissions change'], label='CO2')
plt.plot(df['Pop. change'], label='Pop.')
plt.xlabel("Year", fontsize=20)
plt.ylabel("Change Rates", fontsize=20)
plt.title("Population & CO2 Emissions Change Rate", fontsize=20)
plt.legend(fontsize=15)
```
